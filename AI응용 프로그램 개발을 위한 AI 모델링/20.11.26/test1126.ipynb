```python
import numpy as np
import urllib.request
from keras.utils import to_categorical
```

```python
urllib.request.urlretrieve("http://www.gutenberg.org/files/11/11-0.txt", filename="sample.txt")
f = open('sample.txt', 'rb')
```

```python
lines =[]

for line in f:
    line = line.decode('ascii','ignore')
    line = line.strip()
    line = line.lower()
    if len(line)>0:
        lines.append(line)
        
f.close()
```

```python
lines
```

```python
text = " ".join(lines)
```

```python
print(f'   : {len(text)}')
```

```python
char_vocab = sorted(list(set(text)))
vocab_size = len(char_vocab)
```

```python
# char2idx

char2idx = dict((c,i) for i,c in enumerate(char_vocab))
```

```python
idx2char = {}
for c,i in char2idx.items():
    idx2char[i]=c
```

```python
seq_length = 60
```

```python
n_samples = int(np.floor((len(text) - 1) / seq_length))
```

```python
train_X = []
train_y = []
for i in range(n_samples): 
    X_sample = text[i*seq_length:(i+1)*seq_length]
    
    X_encoded = [char2idx[c] for c in X_sample]
    train_X.append(X_encoded)
    
    Y_sample = text[i*seq_length+1:(i+1)*seq_length+1]
    Y_encoded = [char2idx[c] for c in Y_sample]
    train_y.append(Y_encoded)
```

```python
np.shape(train_X) , np.shape(train_y) 
```

```python
train_X = to_categorical(train_X)
train_y = to_categorical(train_y)
```

```python
train_X.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
model = Sequential()
model.add(LSTM(256, input_shape=(None, train_X.shape[2]), return_sequences=True))
model.add(LSTM(256, return_sequences=True))
model.add(Dense(vocab_size, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(train_X, train_y, epochs=10, verbose=2)
```
