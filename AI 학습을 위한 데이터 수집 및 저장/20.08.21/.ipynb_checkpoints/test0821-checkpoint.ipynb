__ùëâùëéùëüùëñùëéùëõùëêùëí=  1/(ùëõ‚àí1) ‚àë24_(ùëò=1)^ùëõ‚ñí„Äñùë•_ùëò‚àíùë•¬†ÃÖ „Äó__  
__ùë†ùë°ùëéùëõùëëùëéùëüùëë ùëëùëíùë£ùëñùëéùë°ùëñùëúùëõ=‚àöùëâùëéùëüùëñùëéùëõùëêùëí__

__Covariance : 2Ï∞®Ïõê Ï∂ïÏúºÎ°ú , X Î™®ÏñëÏúºÎ°ú sharp ÌïòÎ©¥ Í¥ÄÍ≥ÑÍ∞Ä ÎÜíÏùÄ Îëê Îç∞Ïù¥ÌÑ∞, O Î™®ÏñëÏúºÎ°ú flat ÌïòÎ©¥ Í¥ÄÍ≥ÑÍ∞Ä Ï†ÅÏùÄ Îëê Îç∞Ïù¥ÌÑ∞__

__Í≥µÎ∂ÑÏÇ∞ÏóêÏÑú Î©¥Ï†ÅÏúºÎ°ú Î∂ÑÏÇ∞ Í≥ÑÏÇ∞ Ïãú (x-x_mean)‚óè(y-y_mean) ÏúºÎ°ú => Sum Ïùò Ï†àÎåÄÍ∞íÏù¥ ÌÅ¥ÏàòÎ°ù Í¥ÄÍ≥ÑÍ∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞ Í¥ÄÍ≥Ñ(ÏñëÍ¥ÄÍ≥Ñ,ÏùåÍ¥ÄÍ≥Ñ)__

__ÏÉÅÍ¥Ä Í≥ÑÏàò(Corelation) : ÌëúÏ§ÄÌôîÍ∞Ä Îêú Îëê Î≥ÄÏàòÏùò Í≥µÎ∂ÑÏÇ∞__  
__-1<=corelation<=1 , Ï†àÎåÄÍ∞í Í∏∞Ï§Ä 0.3 Ïñ∏Ï†ÄÎ¶¨ = Í¥ÄÍ≥Ñ ÎÇÆÏùå. 0.7 Ïñ∏Ï†ÄÎ¶¨ = Í¥ÄÍ≥Ñ ÎÜíÏùå__

```python
# AAPL & IBM Í≥µÎ∂ÑÏÇ∞ , ÏÉÅÍ¥ÄÍ≥ÑÏàò Íµ¨Ìï¥Î≥¥Í∏∞

AAPL_mean,IBM_mean,*rest = price.pct_change().mean().values
AAPL_std,IBM_std,*rest2 = price.pct_change().std().values

N = len(price.iloc[0])

Covariance_df = (price.pct_change()['AAPL']-AAPL_mean)*(price.pct_change()['IBM']-IBM_mean)
Covariance = Covariance_df.sum()/(N-1)
print(Covariance)
    
Corelation_df = ((price.pct_change()['AAPL']-AAPL_mean)/AAPL_std)*((price.pct_change()['IBM']-IBM_mean)/IBM_std)
Corelation = Corelation_df.sum()/(N-1)
print(Corelation)
```
